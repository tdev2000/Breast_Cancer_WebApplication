```python
#PHASE 1
#Import Libraries
#---------------------------------------------------------------------------------------
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#PHASE2
#Check working Directory
print("Current directory:")
print(os.getcwd())

print("\nFiles here:")
print(os.listdir())
```

```python
#Explore Dataset Folder
dataset_path = "dataset"
print(os.listdir(dataset_path))
```

```python
#PHASE3
#Load image and Mask safely
image_path = "dataset/benign/benign (1).png"
mask_path = "dataset/benign/benign (1)_mask.png"

image = cv2.imread(image_path)
mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

print("Image loaded:", image is not None)
print("Mask loaded:", mask is not None)
```

```python
#PREPROCESS IMAGE AND MASK
#Visualise original Image+Mask
#Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(mask, cmap="gray")
plt.title("Mask")
plt.axis("off")

plt.show()
```

```python
#Resize Image & Mask
IMG_SIZE = 224

image_resized = cv2.resize(image_rgb, (IMG_SIZE, IMG_SIZE))
mask_resized = cv2.resize(mask, (IMG_SIZE, IMG_SIZE))
```

```python
#Apply Mask
#Convert mask to 0-1 range
mask_binary = mask_resized/ 255.0 #convert to 0-1

#Apply mask to image
masked_image = image_resized * mask_binary[..., np.newaxis]
```

```python
#Visualising masked image
plt.figure(figsize=(5,5))
plt.imshow(masked_image.astype(np.uint8))
plt.title("Masked Ultrasound Image")
plt.axis("off")
plt.show()
```

```python
#phase 5-Preparing the full dataset for training

#----------------------------------------------------------------------------------
#Dataset Loading
#Define constants and labelsis
IMG_SIZE = 224
dataset_path = "dataset"

labels = {
        "normal": 0,
        "benign": 1,
        "malignant": 2
}
```

```python
#Create empty lists

x = []  #images
y = []  #lables
```

```python
#Loop through each class folder
for class_name, class_index in labels.items():
    #Build full path to class folder
    class_path = os.path.join(dataset_path, class_name)
    print(f"Processing folder: {class_name}")
    #Load images and masks
    for file in os.listdir(class_path):
    
        #skip mask files
        if "_mask" in file or not file.endswith(".png"):
            continue
        #Builds Path
        image_path = os.path.join(class_path, file)
        mask_path = image_path.replace(".png", "_mask.png")
    
        image = cv2.imread(image_path)
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    
        if image is None or mask is None:
            continue
            
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image_resized = cv2.resize(image_rgb, (IMG_SIZE,IMG_SIZE))
        mask_resized = cv2.resize(mask, (IMG_SIZE, IMG_SIZE))
        
        mask_binary = mask_resized/255.0
        masked_image = image_resized * mask_binary[..., np.newaxis]

        x.append(masked_image)
        y.append(class_index)
```

```python
#Converting lists to Numpy arrays
x = np.array(x, dtype=np.float32)
y =np.array(y)

print("Dataset shape:", x.shape)
print("Labels shape:", y.shape)
```

```python
#Phase 6
#Train, Test and split data
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,y,
    test_size=0.2,
    random_state=42,
)

print("Training data shape:", x_train.shape)
print("Validation data shape:", x_test.shape )
```

```python
#BUILD CNN
#Import libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
        Conv2D, MaxPooling2D,
        Flatten, Dense, Dropout
)
from tensorflow.keras.optimizers import Adam
```

```python
#Build CNN -Model Architecture
#Create CNN Model
model = Sequential()
#Convolutional Layer-1
model.add(Conv2D(
    filters=32,
    kernel_size=(3, 3),
    activation='relu',
    input_shape=(224,224,3)
))
#Max Pooling Layer
model.add(MaxPooling2D(pool_size=(2,2)))
#Convolution Layer 2
model.add(Conv2D(
    filters=64,
    kernel_size=(3,3),
    activation='relu'
))
#Second MaxpOOLING
model.add(MaxPooling2D(pool_size=(2,2)))
#Flatten layer
model.add(Flatten())
#Dense Layer
model.add(Dense(128, activation='relu'))
#Dropout Layer
model.add(Dropout(0.5))
#Output layer
model.add(Dense(3, activation='softmax'))
```

```python
#Complie the Model
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
#View model summary
model.summary()
```

```python
#Train the model
history = model.fit(
    x_train, y_train,
    validation_data=(x_test, y_test),
    epochs=10,
    batch_size=16
)
```

```python
#Plot tarining graphs
#----------------------------------------------------------------------------------
plt.figure(figsize=(12,4))

#Accuracy
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Testing Accuracy')
plt.legend()
plt.title("Model Accuracy")

#Loss
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.legend()
plt.title("Model Loss")

plt.show()
```

```python
#Evaluate Model on Valdation Data
loss, accuracy = model.evaluate(x_test, y_test)
print("Validation accuracy", accuracy)
```

```python
#Save then model for Flask
model.save("saved_model/breast_cancer_cnn.keras")
```
